```python
def cleaning_data(data):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    for user in data:
        row_rating = user['rating'].strip().lower()
        if (row_rating in text_to_num):
            row_rating = text_to_num[row_rating]
        # print(row_rating)

        user['rating'] = row_rating
        print(user)

cleaning_data(data)
```

```python
# Handle missing values
def cleaning_data(data):
    for user in data:
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = "null"
        print(raw_age)

cleaning_data(data)
```

```python
# Handle Deduplicatioin values
def cleaning_data(data):
    unique_users = set()
    cleaned_data = []
    for user in data:
        name = user["name"].strip()
        if name in unique_users :
            continue

        unique_users.add(name)
        cleaned_data.append(user)

    return cleaned_data
    
new_data = cleaning_data(data)

for user in new_data:
    print(user)
        
```

```python
# meaning data from json file
def important(data):
    total_rating = 0
    for users in data:
        total_rating += float(users["rating"])

    avg = int(total_rating / len(data))

    print(f"Average Rating : {avg}")
```

```python
def poor_rating(data):
    total_rating = 0
    for users in data:
        if float(float(users["rating"]) < 3):
            total_rating += 1

    percent = total_rating / len(data) * 100

    print(f"Percentage of Poor ratting : {percent} %")
```

```python
poor_rating(data)
```
