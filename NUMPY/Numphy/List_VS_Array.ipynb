```python
import numpy as np
import time
```

```python
# execution performance
size = 100_00

# python list
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time = {end-start} seconds")

# numpy array
np_arr = np.array(py_list)
start = time.time()
sq_array = np_arr ** 2
end = time.time()

print(f"python numpy time = {end-start} seconds")
```

```python
# memory 
import sys

print(f"python list size = {sys.getsizeof(py_list)* len(py_list)} bytes")
print(f"numpy array size = {np_array.nbytes} bytes")
```
