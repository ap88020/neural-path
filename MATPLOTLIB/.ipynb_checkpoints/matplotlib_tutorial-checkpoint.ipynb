```python
# read data
oscar_movies = [
    "The dark night",
    "The hurt locker",
    "The king's speech",
    "The artist",
    "Argo"
]

years = [2008,2009,2010,2011,2012]
revenue = [1005,170,427,133,232]

plt.plot(years,revenue)
plt.title("Oscar Movies Revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue (in M$)")
```

```python
# Non-Oscars winners movies
non_oscars_movies = ["slumdon millionar","Avatar","Inception","Hugo","Lincoln"]
non_oscars_years = [2008,2009,2010,2011,2012,]
non_oscars_revenue = [378,2788,829,185,275]

plt.plot(years,revenue, "o--r" ,label="oscars movies")
plt.plot(years,non_oscars_revenue,color="magenta",marker="<",linestyle="--",label="non-oscars movies")
plt.title("Oscars winning revenue VS non Oscars winning revenue")
plt.xlabel("years")
plt.ylabel("Revenue (in M $)")
plt.legend()
```
