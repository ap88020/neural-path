```python
import matplotlib.pyplot as plt
```

```python
# Scatter Plot Data

people = [
    "Person A", "Person B", "Person C", "Person D", "Person E",
    "Person F", "Person G", "Person H", "Person I", "Person J"
]

age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]

bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ["green" if x < 135 else "red" for x in bp]
# plt.scatter(age,bp,cmap="magma",s=bp, marker="<")
plt.scatter(age, bp, c=bp, cmap="magma", s=bp, marker="<")

plt.title("Age VS BP")
plt.xlabel("age")
plt.ylabel("blood pressure")
plt.grid()
plt.colorbar(label="blood pressure")

for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i]+2,bp[i]))
```

```python
# Multiple Scatter Plot Data

cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]


plt.scatter(winter_temp,winter_humidity,label="winter")
plt.scatter(summer_temp,summer_humidity,label="summer")

plt.title("Summer Vs Winter data")
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity (%)")
plt.legend()
plt.grid()
```

```python
expenses = ["Salareis", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode = [0,0,0,0.1,0]
plt.pie(
    amounts,
    labels=expenses,
    autopct="%1.1f%%",
    wedgeprops={
        "edgecolor": "black",
        "linewidth": 1,
        "linestyle": "--",
    },
    explode=explode,
    startangle=90,
    shadow=True
)

plt.show()
```
