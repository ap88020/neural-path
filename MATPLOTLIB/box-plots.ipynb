```python
# Stack Plot Data

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

direct = [50, 60, 70, 80, 90, 100, 110]
organic = [30, 40, 50, 55, 60, 70, 80]
social = [20, 25, 30, 35, 40, 50, 60]

# plt.pie([direct[0],organic[0],social[0]],labels=["direct","organic","social"], autopct="%1.1f%%")
plt.stackplot(days,direct,organic,social,labels=["Direct","Organic","Soical"])
plt.title("Marketing data for the week")
plt.xlabel("Days")
plt.ylabel("# of customers")
plt.legend()
```

```python
# Task Data

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

fig , axes = plt.subplots(2,2)
city_no = 0
for i in range(2):
    for j in range(2):
        axes[i][j].plot(days,temperatures[city_no],marker="o")
        axes[i][j].set_title(cities[city_no])
        axes[i][j].grid(True)
        city_no = city_no + 1

fig.suptitle("Temperature in cities over the week")
fig.supxlabel("days")
fig.supylabel("Tempratures")
fig.tight_layout()
```
