```python
import matplotlib.pyplot as plt
```

```python
# Bar charts
oscar_movies = [
    "The dark night",
    "The hurt locker",
    "The king's speech",
    "The artist",
    "Argo"
]

years = [2008,2009,2010,2011,2012]
revenue = [1005,170,427,133,232]

plt.bar(years,revenue)
plt.title("Revenue in each year for oscars movies")
plt.xlabel("years")
plt.ylabel("revenue in $ M") 

for i in range(len(years)):
    plt.text(years[i],revenue[i]+5,str(revenue[i]),ha="center")
```

```python
# Bar charts
import numpy as np;

oscar_movies = [
    "The dark night",
    "The hurt locker",
    "The king's speech",
    "The artist",
    "Argo"
]

years = [2008,2009,2010,2011,2012]
revenue = [1005,170,427,133,232]

# Non-oscars movies 
non_oscars_movies = ["slumdon millionar","Avatar","Inception","Hugo","Lincoln"]
non_oscars_years = [2008,2009,2010,2011,2012,]
non_oscars_revenue = [378,2788,829,185,275]

x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2,revenue,width,label="Oscars Movies")
plt.bar(x+width/2,non_oscars_revenue,width,label="Oscars Movies")
plt.title("Oscars Vs Non-oscars Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue in $M")
plt.legend()
plt.xticks(x,years)
```
