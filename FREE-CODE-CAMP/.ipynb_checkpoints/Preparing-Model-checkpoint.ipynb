```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cols = ["fLength","fWidth","fSize","fCone","fConel","fAsym","fM3Long","fM3Trans","fAlpha","fDist","class"]
df = pd.read_csv("magic04.data",names=cols)
df.head()
```

```python
df["class"] = (df["class"] == "g").astype(int)
```

```python
df.head()
```

```python
for label in cols[:-1]:
    plt.hist(df[df["class"] == 1][label],color="blue",label="gamma",alpha=0.7,density=0.7)
    plt.hist(df[df["class"] == 0][label],color="red",label="hadron",alpha=0.7,density=0.7)
    plt.title(label)
    plt.xlabel(label)
    plt.ylabel("probability")
    plt.legend()
    plt.show()
```

```python
# train, valid, test = np.split(df.sample(frac=1),[int(0.6*len(df)),int(0.8*len(df))])
from sklearn.model_selection import train_test_split

x = df.drop("class",axis=1)
y = df["class"]

# y.head()
# x.head()

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.6,random_state=42)
```
