```python
# Merging and Joining
df_customers = pd.DataFrame({
    "customer":[1,2,3,4],
    "name":["Adam","Bob","Charlie","Dave"]
})

df_orders = pd.DataFrame({
    "order_id":[101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount":[250,120,300,180]
})
```

```python
# Concatenation
pd1 = pd.DataFrame({
    "id":[1,2,3],
    "name":["Adam","Eve","Bob"]
})

pd2 = pd.DataFrame({
    "id":[4,5,6],
    "name":["steve","banner","tom"]
})

pd.concat([pd1,pd2]) # rows
pd.concat([pd1,pd2],ignore_index=True) # rows
pd.concat([pd1,pd2],axis=1) # col
```
