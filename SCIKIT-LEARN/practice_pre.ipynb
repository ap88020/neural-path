```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split


data = {
    'StudyHours':[1,2,3,4,5],
    'TestScore':[40,50,60,70,80],
}

df = pd.DataFrame(data)

# standerization
standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(df)

print("Stander-Scaler-output")
print(pd.DataFrame(standard_scaled, columns=["Study-Hours","Test-Score"]))
```

```python
# Min-Max-Scaller
minmax_scaller = MinMaxScaler()
minmax_scalled = minmax_scaller.fit_transform(df)

print("Min-Max-Scaller output")
print(pd.DataFrame(minmax_scalled,columns=["Study-Hours","Test-Score"]))
```

```python
# Linear-Regression
from sklearn.linear_model import LinearRegression

x = [[1],[2],[3],[4],[5]]
y = [[10],[20],[30],[40],[50]]

model = LinearRegression()
model.fit(x,y)

hours = float(input("How many hours you studid :"))
predict_marks = model.predict([[hours]])

print(f"A student who studid {hours} got {predict_marks} marks")
```

```python
# Logistics-Regression
from sklearn.linear_model import LogisticRegression

x = [[1],[2],[3],[4],[5]]
y = [[0],[0],[1],[1],[1]]

model = LogisticRegression()
model.fit(x,y)

hours = float(input("Enter how many hours you studid :"))
result = model.predict([[hours]])

if result == 1:
    print(f"You are likely passed based on your {hours} hours")
else:
    print(f"You are likely fail based on your {hours} hours")
```

```python
# K-N-N
from sklearn.neighbors import KNeighborsClassifier

# Weightg & Size
x = [
    [180,7],
    [180,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5],
]

# 0 = Apple & 1 = Orange
y = [0, 0, 1, 1, 1, 1]

model = KNeighborsClassifier()

model.fit(x,y)

weight = float(input("Enter wight in gram"))
size = float(input("Enter size in centimeter"))

prediction = model.predict([[weight,size]])[0]

if prediction == 1:
    print("This is likely to Orange")
else:
    print("This is likely to Apple")
```

```python
from sklearn.tree import DecisionTreeClassifier

x = [
    [7,2], # Apple
    [8,3], # Apple
    [9,8], # orange
    [10,9], # orange
]

y = [0, 0, 1, 1] # 0= Apple , 1= Orange 

model = DecisionTreeClassifier()
model.fit(x,y)

size = float(input("Enter the size of frute : "))
shade = float(input("Enter the color-shade of frute :"))

result = model.predict([[size,shade]])[0]

if result == 0:
    print("this is likley an Apple")
else:
    print("This is likely an Orange")
```
