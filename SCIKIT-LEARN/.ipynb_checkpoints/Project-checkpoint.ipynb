```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error
import numpy as np
```

```python
data = pd.read_csv("student.csv")

data
```

```python
X = data[['Hours']] #double bracket = 2d input
Y = data[['Score']] #target column 

model = LinearRegression()
model.fit(X,Y)

predicted_score = model.predict(X)

# Evaluate
mae = mean_absolute_error(Y,predicted_score)
mse = mean_squared_error(Y,predicted_score)
rmse = np.sqrt(mse)

# show result
print("Mean Absolute Error (MAE) : ", mae)
print("Mean Squared Error (MSE) : ", mse)
print("Root Mean Squared Errr (RMSE) : ", rmse)

new_hour = float(input("Enter new hour : "))
new_pred_score = model.predict([[new_hour]]) 

print(f"predicted score for {new_hour} hour is : {new_pred_score}")
```
