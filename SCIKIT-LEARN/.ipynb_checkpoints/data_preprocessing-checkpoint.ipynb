```python
import pandas as pd
```

```python
data = {
    'Name':['Pawan','Ragiv','Lalit','Ishan','Rohit'],
    'Age':[25,None,44,23,None],
    'Salary':[50000,60000,70000,None,None]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
null_data = df.isnull().sum()
null_data
```

```python
drop_df = df.dropna()
drop_df
```

```python
avg = drop_df['Salary'].mean()
df['salary'].fillna(avg,inplace=True)
df
```
